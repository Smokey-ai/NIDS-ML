```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
def load_nsl_kdd(train_path, test_path=None):
    """
    Load NSL-KDD dataset from CSV format
    """
    print(f"Loading training data from: {train_path}")
    
    # Load training data
    train_df = pd.read_csv(train_path)
    
    # Clean column names (remove any extra spaces)
    train_df.columns = train_df.columns.str.strip()
    
    # Load test data if provided
    test_df = None
    if test_path:
        print(f"Loading test data from: {test_path}")
        test_df = pd.read_csv(test_path)
        test_df.columns = test_df.columns.str.strip()
    
    # Display column information
    print("Columns in dataset:", list(train_df.columns))
    
    # Check if the dataset has the expected structure
    expected_features = 41  # NSL-KDD should have 41 features + 1 label
    if len(train_df.columns) != expected_features + 1:
        print(f"Warning: Expected {expected_features + 1} columns, but found {len(train_df.columns)}")
        print("Please verify your dataset structure")
    
    return train_df, test_df

# Step 2: Explore the dataset
def explore_dataset(df):
    """
    Basic exploration of the dataset
    """
    print("Dataset Shape:", df.shape)
    print("\nFirst 5 rows:")
    print(df.head())
    
    print("\nDataset Info:")
    print(df.info())
    
    print("\nMissing Values:")
    print(df.isnull().sum())
    
    # Find the label column
    possible_label_cols = ['label', 'class', 'target', 'attack_type']
    label_col = None
    
    for col in possible_label_cols:
        if col in df.columns:
            label_col = col
            break
    
    if label_col is None:
        # Assume the last column is the label
        label_col = df.columns[-1]
    
    print(f"\nLabel Distribution (using column '{label_col}'):")
    label_counts = df[label_col].value_counts()
    print(label_counts)
    
    # Plot label distribution
    plt.figure(figsize=(15, 6))
    plt.subplot(1, 2, 1)
    if len(label_counts) > 20:  # Too many classes for readable bar plot
        plt.figure(figsize=(20, 8))
        label_counts.head(20).plot(kind='bar')
        plt.title('Top 20 Attack Types Distribution')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    else:
        label_counts.plot(kind='bar')
        plt.title('Attack Type Distribution')
        plt.xticks(rotation=45)
        
        plt.subplot(1, 2, 2)
        # Group by attack categories
        attack_categories = categorize_attacks(df[label_col].unique())
        category_counts = {}
        for label in df[label_col].unique():
            category = attack_categories.get(label, 'Unknown')
            category_counts[category] = category_counts.get(category, 0) + (df[label_col] == label).sum()
        
        pd.Series(category_counts).plot(kind='pie', autopct='%1.1f%%')
        plt.title('Attack Categories Distribution')
        plt.tight_layout()
        plt.show()
    
    return label_counts, label_col

# Step 3: Categorize attacks
def categorize_attacks(labels):
    """
    Categorize individual attacks into main categories
    """
    dos_attacks = ['neptune', 'smurf', 'pod', 'teardrop', 'land', 'back']
    probe_attacks = ['satan', 'ipsweep', 'nmap', 'portsweep', 'saint', 'mscan']
    r2l_attacks = ['warezclient', 'imap', 'ftp_write', 'guess_passwd', 'multihop', 
                   'warezmaster', 'xlock', 'xsnoop', 'snmpguess', 'snmpgetattack', 
                   'httptunnel', 'sendmail', 'named']
    u2r_attacks = ['buffer_overflow', 'loadmodule', 'perl', 'rootkit', 'ps', 
                   'sqlattack', 'xterm']
    
    attack_categories = {}
    for label in labels:
        if label == 'normal':
            attack_categories[label] = 'Normal'
        elif label in dos_attacks:
            attack_categories[label] = 'DoS'
        elif label in probe_attacks:
            attack_categories[label] = 'Probe'
        elif label in r2l_attacks:
            attack_categories[label] = 'R2L'
        elif label in u2r_attacks:
            attack_categories[label] = 'U2R'
        else:
            attack_categories[label] = 'Unknown'
    
    return attack_categories

# Step 4: Clean the data
def clean_data(df):
    """
    Clean the dataset
    """
    print("Cleaning data...")
    
    # Remove duplicates
    initial_shape = df.shape
    df = df.drop_duplicates()
    print(f"Removed {initial_shape[0] - df.shape[0]} duplicate rows")
    
    # Handle missing values (if any)
    if df.isnull().sum().sum() > 0:
        print("Handling missing values...")
        # For numerical columns, fill with median
        numerical_cols = df.select_dtypes(include=[np.number]).columns
        df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())
        
        # For categorical columns, fill with mode
        categorical_cols = df.select_dtypes(include=['object']).columns
        categorical_cols = categorical_cols.drop('label')  # Don't fill label column
        for col in categorical_cols:
            df[col] = df[col].fillna(df[col].mode()[0])
    
    return df

# Step 5: Feature Engineering
def feature_engineering(df):
    """
    Perform feature engineering
    """
    print("Performing feature engineering...")
    
    # Find the label column (it might be named differently in CSV)
    possible_label_cols = ['label', 'class', 'target', 'attack_type']
    label_col = None
    
    for col in possible_label_cols:
        if col in df.columns:
            label_col = col
            break
    
    if label_col is None:
        # Assume the last column is the label
        label_col = df.columns[-1]
        print(f"Warning: Using last column '{label_col}' as label column")
    
    print(f"Using '{label_col}' as label column")
    
    # Separate features and labels
    X = df.drop(label_col, axis=1)
    y = df[label_col]
    
    # Identify categorical columns (typically object/string types)
    categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
    
    # Common NSL-KDD categorical columns (if present)
    common_categorical = ['protocol_type', 'service', 'flag']
    categorical_cols = [col for col in common_categorical if col in X.columns] + \
                      [col for col in categorical_cols if col not in common_categorical]
    
    numerical_cols = [col for col in X.columns if col not in categorical_cols]
    
    print(f"Found {len(categorical_cols)} categorical columns: {categorical_cols}")
    print(f"Found {len(numerical_cols)} numerical columns")
    
    # Encode categorical variables
    label_encoders = {}
    for col in categorical_cols:
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col].astype(str))  # Convert to string first
        label_encoders[col] = le
        print(f"  Encoded '{col}': {len(le.classes_)} unique values")
    
    return X, y, label_encoders

# Step 6: Scale features
def scale_features(X_train, X_test=None, method='standard'):
    """
    Scale numerical features
    """
    print(f"Scaling features using {method} scaling...")
    
    if method == 'standard':
        scaler = StandardScaler()
    elif method == 'minmax':
        scaler = MinMaxScaler()
    else:
        raise ValueError("Method must be 'standard' or 'minmax'")
    
    # Fit scaler on training data
    X_train_scaled = scaler.fit_transform(X_train)
    X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
    
    # Transform test data if provided
    X_test_scaled = None
    if X_test is not None:
        X_test_scaled = scaler.transform(X_test)
        X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
    
    return X_train_scaled, X_test_scaled, scaler

# Step 7: Encode labels for multi-class classification
def encode_labels(y_train, y_test=None, use_categories=False):
    """
    Encode labels for classification
    """
    print("Encoding labels...")
    
    if use_categories:
        # Group attacks into categories
        attack_categories = categorize_attacks(y_train.unique())
        y_train = y_train.map(attack_categories)
        if y_test is not None:
            y_test = y_test.map(attack_categories)
    
    # Encode labels
    label_encoder = LabelEncoder()
    y_train_encoded = label_encoder.fit_transform(y_train)
    
    y_test_encoded = None
    if y_test is not None:
        y_test_encoded = label_encoder.transform(y_test)
    
    print(f"Number of classes: {len(label_encoder.classes_)}")
    print(f"Classes: {label_encoder.classes_}")
    
    return y_train_encoded, y_test_encoded, label_encoder

# Step 8: Complete preprocessing pipeline
def preprocess_nsl_kdd(train_path, test_path=None, use_categories=False, scaling_method='standard'):
    """
    Complete preprocessing pipeline
    """
    print("Starting NSL-KDD preprocessing pipeline...")
    print("="*50)
    
    # Load data
    print("Step 1: Loading data...")
    train_df, test_df = load_nsl_kdd(train_path, test_path)
    
    # Explore data
    print("\nStep 2: Exploring data...")
    label_counts, label_col = explore_dataset(train_df)
    
    # Clean data
    print("\nStep 3: Cleaning data...")
    train_df = clean_data(train_df)
    if test_df is not None:
        test_df = clean_data(test_df)
    
    # Feature engineering
    print("\nStep 4: Feature engineering...")
    X_train, y_train, label_encoders = feature_engineering(train_df)
    
    X_test, y_test = None, None
    if test_df is not None:
        X_test, y_test, _ = feature_engineering(test_df)
        # Apply same encoding to test data
        for col, le in label_encoders.items():
            X_test[col] = le.transform(X_test[col])
    
    # Scale features
    print("\nStep 5: Scaling features...")
    X_train_scaled, X_test_scaled, scaler = scale_features(X_train, X_test, scaling_method)
    
    # Encode labels
    print("\nStep 6: Encoding labels...")
    y_train_encoded, y_test_encoded, label_encoder = encode_labels(y_train, y_test, use_categories)
    
    print("\nPreprocessing completed!")
    print(f"Training data shape: {X_train_scaled.shape}")
    if X_test_scaled is not None:
        print(f"Test data shape: {X_test_scaled.shape}")
    
    return {
        'X_train': X_train_scaled,
        'X_test': X_test_scaled,
        'y_train': y_train_encoded,
        'y_test': y_test_encoded,
        'label_encoder': label_encoder,
        'feature_encoders': label_encoders,
        'scaler': scaler
    }

# Example usage:
if __name__ == "__main__":
    # Replace with your actual CSV file paths
    train_path = "kdd_train_clean.csv"  # Your training CSV file
    test_path = "kdd_test_clean.csv"    # Your test CSV file (optional)
    
    # Alternative common names for NSL-KDD CSV files:
    # train_path = "KDDTrain+.csv" 
    # test_path = "KDDTest+.csv"
    
    # Preprocess the data
    processed_data = preprocess_nsl_kdd(
        train_path=train_path,
        test_path=test_path,
        use_categories=False,  # Set to True if you want to group attacks into categories
        scaling_method='standard'  # or 'minmax'
    )
    
    # Access processed data
    X_train = processed_data['X_train']
    y_train = processed_data['y_train']
    X_test = processed_data['X_test']
    y_test = processed_data['y_test']
    
    print("\nClass distribution in training data:")
    unique, counts = np.unique(y_train, return_counts=True)
    class_distribution = dict(zip(unique, counts))
    
    # Get class names from label encoder
    label_encoder = processed_data['label_encoder']
    for class_id, count in class_distribution.items():
        class_name = label_encoder.inverse_transform([class_id])[0]
        print(f"Class {class_id} ({class_name}): {count} samples")
    
    print(f"\nTotal samples: {len(y_train)}")
    print(f"Features shape: {X_train.shape}")
    print(f"Number of features: {X_train.shape[1]}")
    print(f"Number of classes: {len(np.unique(y_train))}")
    
    # Save processed data (optional)
    # processed_data['X_train'].to_csv('X_train_processed.csv', index=False)
    # pd.DataFrame(processed_data['y_train'], columns=['label']).to_csv('y_train_processed.csv', index=False)
```
